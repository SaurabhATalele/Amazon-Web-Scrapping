```python
#import rewuired libraries
import time
import json
import pandas as pd
from bs4 import BeautifulSoup
import urllib3
from selenium import webdriver
```

```python
#Read the dataset

df = pd.read_csv("Amazon Scraping.csv")

df.head(25)
```

```python
data_scraped = {}
_404s =["Page introuvable",
        "Amazon.fr page non trouv√©e",
        "Amazon.de Die Website wurde nicht gefunden",
        "Amazon.it Documento non trovato",
       "404 - Documento non trovato",
       "404 - Documento no encontrado"]

```

```python
driver = webdriver.Chrome("chromedriver")

for id,row in df.iterrows():
    start = time.time()
    country = row["country"]
    Asin = row["Asin"]
    link = 'https://www.amazon.'+country+'/dp/'+Asin
    

    driver.get(link)
    r =driver.page_source

    Soup_obj = BeautifulSoup(r,"html.parser")
    #check if page exists
    

    
    if Soup_obj.title.text in _404s  :
        
        print(f'{link} not available')


    else:
        print(link)
        if Soup_obj.find("span",{"class":"a-price-whole"}) == None :
            
            if Soup_obj.find("span",{"class":"a-color-base"}) == None:
                
                Price = Soup_obj.find("span",{"class":"a-color-price"}).get_text()
            else:
                
                Price = Soup_obj.find("span",{"class":"a-color-base"}).get_text()
 
        else:
            Price=Soup_obj.find("span",{"class":"a-price-whole"}).get_text()+Soup_obj.find("span",{"class":"a-price-fraction"}).get_text()
        data_scraped[row['id']]={
            'Product_name':Soup_obj.find("span",{"id":"productTitle"}).get_text(),
            'Img_url':Soup_obj.find("img")['src'],
            'Price':Price,

        }
        

    if(row["Unnamed: 0"]%100 == 99):
        end = time.time()
        print(f'Completed 100 iterations in {end-start}')
```

```python

data = json.dumps(data_scraped)
f = open("details.json",'w')
f.write(data)
f.close()
```
